```python
policy_management_system/
│
├── policyholder.py
├── product.py
├── payment.py
├── main.py
└── README.md

class Policyholder:
    def __init__(self, policyholder_id, name):
        self.policyholder_id = policyholder_id
        self.name = name
        self.status = "Active"
        self.products = []
        self.balance = 0.0

    def register(self):
        self.status = "Active"

    def suspend(self):
        self.status = "Suspended"

    def reactivate(self):
        self.status = "Active"

    def add_product(self, product):
        self.products.append(product)

    def display_details(self):
        print(f"Policyholder ID: {self.policyholder_id}")
        print(f"Name: {self.name}")
        print(f"Status: {self.status}")
        print(f"Products: {[product.name for product in self.products]}")
        print(f"Outstanding Balance: ₦{self.balance}")
        print("-" * 40)

class Product:
    def __init__(self, product_id, name, price):
        self.product_id = product_id
        self.name = name
        self.price = price
        self.active = True

    def update_product(self, name=None, price=None):
        if name:
            self.name = name
        if price:
            self.price = price

    def suspend_product(self):
        self.active = False

    def activate_product(self):
        self.active = True

class Payment:
    def __init__(self, policyholder):
        self.policyholder = policyholder

    def process_payment(self, amount):
        self.policyholder.balance -= amount
        print(f"Payment of ₦{amount} processed for {self.policyholder.name}")

    def send_reminder(self):
        if self.policyholder.balance > 0:
            print(f"Reminder: {self.policyholder.name} owes ₦{self.policyholder.balance}")

    def apply_penalty(self, penalty_amount):
        self.policyholder.balance += penalty_amount
        print(f"Penalty of ₦{penalty_amount} applied to {self.policyholder.name}")


from policyholder import Policyholder
from product import Product
from payment import Payment

# Create Products
health_insurance = Product("P001", "Health Insurance", 50000)
auto_insurance = Product("P002", "Auto Insurance", 30000)

# Create Policyholders
policyholder1 = Policyholder("PH001", "Oruh Precious")
policyholder2 = Policyholder("PH002", "Bryan Oruh")

# Assign Products
policyholder1.add_product(health_insurance)
policyholder2.add_product(auto_insurance)

# Set balances
policyholder1.balance = health_insurance.price
policyholder2.balance = auto_insurance.price

# Process Payments
payment1 = Payment(policyholder1)
payment2 = Payment(policyholder2)

payment1.process_payment(50000)
payment2.process_payment(30000)

# Display Policyholder Details
policyholder1.display_details()
policyholder2.display_details()


## Author
Oruh Precious


```
